```python
# The purpose of this program is to take a file entered by the user and then saves
# the data from the file into a list. It then takes the list and graphs the data 
# in both a line chart and a bar chart. They are seperated by year and by profit.

# Author: Aidan Pavlik



import matplotlib.pyplot as plt
import os

filename = input("Enter File Path").strip()
yCords = []
xCords = []



if os.path.isfile(filename):
    infile = open(filename, "r")

    line = infile.readline()
    while line != "" :
        splitLine = line.split()
        t1 = (splitLine[0])
        t2 = splitLine[2]
        t2  =  t2[1:]
        t2 = t2.replace(",", "")
        
        print(t1 + " " + t2)
        
        if t2 != "rofit":
            t2 = int(t2)
            xCords.append(t1)
            yCords.append((t2))

        #print(splitLine[0], " ", splitLine[1], (t1+t2)/2)

        line = infile.readline()

    plt.plot(xCords, yCords)
    
    plt.title("Company's Profit Over the Past 10 Years")
    plt.xlabel('Years')
    plt.ylabel('Net Profits in the 100 Millions')
    plt.grid(True)

    
    plt.show()
    
    plt.bar(xCords,yCords)
    plt.title("Company's Profit Over the Past 10 Years")
    plt.xlabel('Years')
    plt.ylabel('Net Profits in the 100 Millions')
    plt.show
    
    
    
    
    infile.close()
else:
    print("Failed to open File!")
```

```python
# This programs purpose is to take a file entered by a user and graph the contents of
# the data into a piechart. The piechart is then seperated by deoeartment and the numer of employees 
# in each department. It also shows the percentage.

# Author: Aidan Pavlik



import matplotlib.pyplot as plt
import os

filename = input("Enter File Path").strip()
yCords = []
xCords = []



if os.path.isfile(filename):
    infile = open(filename, "r")

    line = infile.readline()
    while line != "" :
        splitLine = line.split(", ")
        t1 = (splitLine[0])
        t2 = splitLine[1]
        
        
        
        print(t1 + " " + t2)
        
        if t1 != "Department Name       ":
            #t2 = int(t2)
            xCords.append(t1)
            yCords.append((t2))

        #print(splitLine[0], " ", splitLine[1], (t1+t2)/2)

        line = infile.readline()

    
    
    plt.pie(yCords, labels=xCords, autopct='%1.1f%%')
    plt.title("Employee by Department")
    plt.show

    
    
    
    
    
    infile.close()
else:
    print("Failed to open File!")








```
