```python
#This program is designed to how the different functions of a pandas. Specifically how dataframes and Series Work
#Author: Aidan Pavlik


import random

import pandas as pd

li = [7, 11, 13, 17]
randList = []

ser = pd.Series(li)
ds = pd.Series([100, 100, 100, 100, 100])
randIntSeries = pd.Series()

for i in range(20):
    randList.append(random.randint(1,100))
randIntSeries = pd.Series(randList)

tempature = pd.Series([98.6, 98.9, 100.2, 97.9], index= ["Julie", "Charlie", "Sam", "Andrea"])

tempDict = {
  "Julie": 98.6,
  "Charlie": 98.9,
  "Same": 100.2,
  "Andrea": 97.9
}

tempDictSer = pd.Series(tempDict)


print(ser)
print("\n")
print(ds)
print("\n")
print(randIntSeries)
print(randIntSeries.describe())
print("\n")
print(tempature)
print("\n")
print(tempDictSer)
print("\n")
print("-------------------------------------------------------------------------------------")
print("\n")


tempDict = {
  "Maxine": [97.6, 100.8, 103.2],
  "James": [95.6, 99.4, 101.2],
  "Amanda": [102.4, 99.3, 98.6]
}

tempDataFrame = pd.DataFrame(tempDict)

indList = ["Morning", "Afternoon", "Evening"]

indListIndex = pd.DataFrame(tempDict, index = indList)

print(tempDataFrame)
print("\n")
print(indListIndex)
print("\n")
print(indListIndex[["Maxine"]])
print("\n")
print(indListIndex.loc[["Morning"]])
print("\n")
print(indListIndex.loc[["Morning", "Evening"]])
print("\n")
print(indListIndex[["Maxine", "Amanda"]])
print("\n")
print(indListIndex.loc[["Morning", "Afternoon"], ["Maxine", "Amanda"]])
print("\n")
print(indListIndex.describe())
print("\n")
print(indListIndex.transpose())
print("\n")
print(indListIndex[sorted(indListIndex.columns)])
```

```python
#This program takes an external file, in this case titanic and uses pandas to datamine various different stats about it
#It then saves all of those stats into a new file called savedData.txt
#Author: Aidan Pavlik

import csv
import pandas as pd

data = pd.read_csv('titanic.csv')
tiDataFrame = pd.DataFrame(data)
print("Number of passengers: ", len(tiDataFrame.index))
print("Number of Female Passengers: ", len(tiDataFrame[tiDataFrame["sex"] == "female"]))
print("Number of Male Passengers: ", len(tiDataFrame[tiDataFrame["sex"] == "male"]))
print("Mean average age of passengers: ", tiDataFrame["age"].mean())
print("Number of Passengers under the age of 21: ", len(tiDataFrame[tiDataFrame["age"] < 21]))
print("Number of Passengers to survive: ", len(tiDataFrame[tiDataFrame["survived"] == "yes"]))
print("Number of male Passengers to survive: ",
      len(tiDataFrame[(tiDataFrame["sex"] == "male") & (tiDataFrame["survived"] == "yes")]))
print("Number of female Passengers to survive: ",
      len(tiDataFrame[(tiDataFrame["sex"] == "female") & (tiDataFrame["survived"] == "yes")]))

younSurRow = tiDataFrame.loc[(tiDataFrame['survived'] == 'yes') & (tiDataFrame['age'] == tiDataFrame.loc[tiDataFrame['survived'] == 'yes', 'age'].min())]
younSurName = younSurRow['Name'].iloc[0]
younSurAge = younSurRow['age'].iloc[0]
print("Yougest age to survive: " , younSurAge, " Name: ", younSurName)

oldSurRow = tiDataFrame.loc[(tiDataFrame['survived'] == 'yes') & (tiDataFrame['age'] == tiDataFrame.loc[tiDataFrame['survived'] == 'yes', 'age'].max())]
oldSurName = oldSurRow['Name'].iloc[0]
oldSurAge = oldSurRow['age'].iloc[0]
print("Oldest Age to survive: " , oldSurAge, " Name: ", oldSurName)

surNames = tiDataFrame.loc[tiDataFrame['survived'] == 'yes', 'Name']
print("List of passengers: \n", surNames)


message = " "

message += str(("Number of passengers: ", len(tiDataFrame.index)))
message +="\n" + str(("Number of Female Passengers: ", len(tiDataFrame[tiDataFrame["sex"] == "female"])))
message +="\n" + str(("Number of Male Passengers: ", len(tiDataFrame[tiDataFrame["sex"] == "male"])))
message +="\n" + str(("Mean average age of passengers: ", tiDataFrame["age"].mean()))
message +="\n" + str((("Number of Passengers under the age of 21: ", len(tiDataFrame[tiDataFrame["age"] < 21]))))
message +="\n" + str(("Number of Passengers to survive: ", len(tiDataFrame[tiDataFrame["survived"] == "yes"])))
message +="\n" + str(("Number of male Passengers to survive: ",
       len(tiDataFrame[(tiDataFrame["sex"] == "male") & (tiDataFrame["survived"] == "yes")])))
message +="\n" + str(("Number of female Passengers to survive: ",
       len(tiDataFrame[(tiDataFrame["sex"] == "female") & (tiDataFrame["survived"] == "yes")])))
message +="\n" + str(("Yougest age to survive: " , younSurAge, " Name: ", younSurName))
message +="\n" + str(("Oldest Age to survive: " , oldSurAge, " Name: ", oldSurName))
message +="\n" + str(("List of passengers: \n", surNames))





file = open('savedData.txt', 'w')

file.write(message)



file.close()

```
