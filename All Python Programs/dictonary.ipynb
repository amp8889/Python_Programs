```python
import pickle
import json

userDict = {'mike': 'mike@gmail.com', 'jake': 'jake@yahoo.com'}

with open("dictData1.txt", "r") as fp:
    # Load the dictionary from the file
    userDict = json.load(fp)


print("1. Add a new name and email address \n2. Find an existing email address from a name. \n3. Change an existing email address. \n4. Delete an existing name and email address. \n5. Exit the program\n")
numEntered = input("Enter the number of the action you wish to do: ")


while numEntered != 5:
    if numEntered == "1":
        print("1 entered")
        new_key = input("Please enter the name: ")
        new_value = input("Please enter the email: ")
        userDict = {**userDict, new_key: new_value}
        print("The name and email has been added\n")
    elif numEntered == "2":
        nameEntered = input("Enter the name you wish to find: ")
        print(userDict[nameEntered])
        print("")

    elif numEntered == "3":
        nameEntered = input("Enter the name you wish to change: ")
        newEmail = input("Enter the new email address: ")
        userDict[nameEntered] = newEmail
        print("The email has been updated\n")

    elif numEntered == "4":
        nameEntered = input("Enter the name you wish to delete: ")
        userDict.pop(nameEntered, None)
        print("The name and email has been deleted\n")

    elif numEntered == "5":
        break
    
    numEntered = input("Enter the number of the action you wish to do: ")







print(userDict)

    
    

 


with open('dictData1.txt', 'w') as file:
     file.write(json.dumps(userDict)) # use `json.loads` to do the reverse




















```
