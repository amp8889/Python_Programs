```python
#This program creates a GUI table that allows the user to enter a certain number of coins and then
#will determine the value of the coins. It will also prevent invalid inputs from being used
#Author: Aidan Pavlik

from tkinter import *
window = Tk()
window.title("Welcome Money Calculator")
window.geometry('400x400')
lblPennie = Label(window, text="Pennies")
lblPennie.grid(column=0, row=0)

lblNickle = Label(window, text="Nickles")
lblNickle.grid(column=0, row=1)

lblDime = Label(window, text="Dimes")
lblDime.grid(column=0, row=2)

lblQuarter = Label(window, text="Quarters")
lblQuarter.grid(column=0, row=3)

lblHalfDollar = Label(window, text="Half Dollar")
lblHalfDollar.grid(column=0, row=4)

lblDollarCoin = Label(window, text="Dollar Coin")
lblDollarCoin.grid(column=0, row=5)

lblTotal = Label(window, text="Total")
lblTotal.grid(column=0, row=6)

lblTotalCoin = Label(window, text="")
lblTotalCoin.grid(column=1, row=6)

txtPennie = Entry(window,width=10)
txtPennie.grid(column=1, row=0)

txtNickle = Entry(window,width=10)
txtNickle.grid(column=1, row=1)

txtDime = Entry(window,width=10)
txtDime.grid(column=1, row=2)

txtQuarter = Entry(window,width=10)
txtQuarter.grid(column=1, row=3)

txtHalfDollar = Entry(window,width=10)
txtHalfDollar.grid(column=1, row=4)

txtDollarCoin = Entry(window,width=10)
txtDollarCoin.grid(column=1, row=5)

def clicked():


    try:

        total = (int(txtPennie.get()) * .01) + (int(txtNickle.get()) * .05) + (int(txtDime.get()) * .10) + (int(txtQuarter.get()) * .25) + (int(txtHalfDollar.get()) * .50) + (int(txtDollarCoin.get()) * 1)
        total = round(total, 2)
        if total < 0:
            lblTotalCoin.configure(text="Invalid Input Please fix your coins")
        else:
            lblTotalCoin.configure(text= total)
    except:
        lblTotalCoin.configure(text= "Invalid Input Please fix your coins")



btn = Button(window, text="Get Total", command=clicked)
btn.grid(column=4, row=8)
window.mainloop()
```

```python
# This program checks if a password is valid based on a condition, in this case, if it contains at least one digit
# Author: Aidan Pavlik

def DoesPasswordPassCheck(password):
    try:
        if len(password) == 0:
            return False
        elif password[0].isdigit():
            return True
        else:
            return DoesPasswordPassCheck(password[1:])
    except:
        return False


def main():
    # Invalid Inputs
    print(DoesPasswordPassCheck("Password"))
    print(DoesPasswordPassCheck("@id*n"))
    print(DoesPasswordPassCheck("this is not a valid password"))
    # Valid Inputs
    print(DoesPasswordPassCheck("Passw0rd"))
    print(DoesPasswordPassCheck("P1234Wo1d"))
    print(DoesPasswordPassCheck("@1d@n"))


main()
```
